```python
import numpy as np
import matplotlib.pyplot as plt
# Line chart

t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2.5*np.pi*t)
plt.plot(t,s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2
print(linear_data, exponential_data)
plt.plot(linear_data, '-o', exponential_data, '-x')
plt.gca().fill_between(range(len(linear_data)),
                       linear_data, exponential_data,
                       facecolor='blue',
                       alpha=0.25)
plt.legend(['Linear line', 'Exponential line'])
plt.show()
```

```python
import pandas as pd
observation_dates = np.arange('2017-01-01', '2017-01-09', 	dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, 	observation_dates))
plt.plot(observation_dates, linear_data, '-o',  	observation_dates, exponential_data, '-o')


plt.legend(['Linear line', 'Exponential line'])
x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)
# adjust the subplot so the text doesn't run off the image
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```python
xvals = range(len(linear_data))
print(xvals)
plt.bar(xvals, linear_data, width=0.3)

exp_xvals = []
for item in xvals:
    exp_xvals.append(item+0.3)
plt.bar(exp_xvals, exponential_data, width=0.3, color='r')


plt.legend(['Linear data', 'Exponential data'])
plt.show()
```

```python
temperatures = np.array([-195, 0, 20, 100])
mean_impact = np.array([1.4, 62.2, 70.4, 77.4])
std_error = np.array([0.2, 9.3, 6.5, 1.9])
fig, ax = plt.subplots()
ax.errorbar(temperatures, mean_impact, yerr=std_error, fmt='--o')
ax.set_xlabel('Temperature')
ax.set_ylabel('Mean of Impact Energy')
plt.show()
```

```python
# Scatter plot
# Create data
N = 500
x = np.random.rand(N)
y = np.random.rand(N)
colors = (1, 0.5, 0)
area = np.pi * 3

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
